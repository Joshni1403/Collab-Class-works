#TASK 1
Read the sentences and label them manually as: Positive / Negative / Neutral



```python
import pandas as pd

data = {
    "text": [
        "Woke up before the alarm and somehow felt human — rare, but glorious.",
        "It wasn’t a disaster, just… consistently disappointing.",
        "Sure, autocorrect, change everything I type — you clearly know me better.",
        "The phone battery dropped to 50% by noon.",
        "Oh amazing — another “urgent” task that magically appears at 5:59 PM",
        "My energy vanished faster than my weekend plans.",
        "That recipe experiment actually tasted restaurant-level good!",
        "Wonderful, more economic indicators — because spreadsheets and uncertainty are obviously fun weekend plans."
    ]
}

df = pd.DataFrame(data)
print (df)
```

```python
#Process 2
import nltk
nltk.download('vader_lexicon')

from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

df['score'] = df['text'].apply(lambda x: sia.polarity_scores(x)['compound'])
df['sentiment'] = df['score'].apply(lambda x:
                                    'Positive' if x > 0
                                    else ('Negative' if x < 0
                                    else 'Neutral'))
df
```

```python
#process 3
!pip install afinn
from afinn import Afinn

af = Afinn()
df['score'] = df['text'].apply(lambda x: af.score(x))
df['sentiment']
```

```python
#process 4
!pip install textblob
from textblob import TextBlob
for i in df['text']:
  print(TextBlob(i).sentiment)
```

```python
#Let us compare !
from textblob import TextBlob

df['textblob_score'] = df['text'].apply(lambda x: TextBlob(x).sentiment.polarity)
df['textblob_sentiment'] = df['textblob_score'].apply(
    lambda x: 'Positive' if x>0 else ('Negative' if x<0 else 'Neutral')
)
df
```
